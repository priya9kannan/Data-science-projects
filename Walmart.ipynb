```python
avgamt_male=user_total[user_total['Gender']=='M']
avgamt_female=data=user_total[user_total['Gender']=='F']
```

```python
#finding purchase amount distribution for males & females 

sample_size = 1000

num_rep=1000

male_means = []
female_means=[]

for i in range(num_rep):    
    male_mean=avgamt_male.sample(sample_size,replace=True)['Purchase'].mean()
    female_mean=avgamt_female.sample(sample_size,replace=True)['Purchase'].mean()
    
    male_means.append(male_mean)
    female_means.append(female_mean)
    
```

```python
#fig,axs=plt.subplots(nrows=1,ncols=2,figsize=(15,9))

#sns.histplot(data=male_means,ax=axs[0]).set_title("Purchase_amount distribution by males for 1000 samples")
#sns.histplot(data=female_means,ax=axs[1]).set_title("Purchase_amount distribution by females 1000 samples")


sns.histplot(data=male_means,color='b')
sns.histplot(data=female_means)
```

The mean value of samples for both males and females are same for a sample of 1000.

```python
#Calculating 90% confidence interval for sample size 1000:

z= 1.645

sample_men=np.mean(male_means)
sample_women=np.mean(female_means)

stdsample_men=pd.Series(male_means).std()
stdsample_women=pd.Series(female_means).std()

stderr_men = stdsample_men/np.sqrt(1000)
stderr_fem=stdsample_women/np.sqrt(1000)

upper_m = sample_men+(z * stderr_men)
upper_f = sample_women+(z* stderr_fem)

lower_m = sample_men-(z * stderr_men)
lower_f = sample_women-(z* stderr_fem)

print("Male C.I",[lower_m,upper_m])
print("Female C.I",[lower_f,upper_f])
```

From above observations,we can say that for 90% CI

Avg amount spent on Female C.I [709678.5208199603, 712401.3197540395]
